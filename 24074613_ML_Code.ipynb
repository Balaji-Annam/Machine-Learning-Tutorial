```python
# ============================================================
# 1. Imports and Setup
# ============================================================

import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import itertools

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow version:", tf.__version__)


# ============================================================
# 2. Load and Preprocess CIFAR-10 Data
# ============================================================

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
y_train = y_train.flatten()
y_test = y_test.flatten()

print("Train shape:", x_train.shape, y_train.shape)
print("Test shape:", x_test.shape, y_test.shape)

# Normalize to [0,1]
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

num_classes = 10
class_names = [
    "airplane", "automobile", "bird", "cat", "deer",
    "dog", "frog", "horse", "ship", "truck"
]

# One-hot encode labels for Keras (only for training)
y_train_cat = to_categorical(y_train, num_classes)
y_test_cat = to_categorical(y_test, num_classes)


# ============================================================
# 3. Define Data Augmentation Strategies
# ============================================================

# Baseline: no augmentation, only rescaling (already normalized above)
baseline_datagen = ImageDataGenerator()

# Simple horizontal flip
flip_datagen = ImageDataGenerator(
    horizontal_flip=True
)

# Rotations
rotation_datagen = ImageDataGenerator(
    rotation_range=20
)

# Shifts and zoom (light geometric augmentation)
shift_zoom_datagen = ImageDataGenerator(
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)

# Combined "heavier" augmentation: rotation + shift + zoom + flip
heavy_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

augmentations = {
    "baseline": baseline_datagen,
    "flip": flip_datagen,
    "rotation": rotation_datagen,
    "shift_zoom": shift_zoom_datagen,
    "heavy": heavy_datagen
}


# ============================================================
# 4. Define a CNN Model Builder (Same Architecture for All)
# ============================================================

def build_cnn_model(input_shape=(32, 32, 3), num_classes=10):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation="relu", padding="same", input_shape=input_shape),
        layers.Conv2D(32, (3, 3), activation="relu", padding="same"),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Conv2D(64, (3, 3), activation="relu", padding="same"),
        layers.Conv2D(64, (3, 3), activation="relu", padding="same"),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax")
    ])

    model.compile(
        optimizer="adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


# ============================================================
# 5. Train CNN for Each Augmentation Setting
# ============================================================

BATCH_SIZE = 64
EPOCHS = 15
VAL_SPLIT = 0.2

histories = {}
models_dict = {}
results = {}

for name, datagen in augmentations.items():
    print(f"\n==============================")
    print(f"Training model with augmentation: {name}")
    print(f"==============================")

    # Build a fresh model
    model = build_cnn_model()
    models_dict[name] = model

    # Create training generator from x_train, y_train
    # We use flow() with validation_split manually by splitting arrays.
    num_train = int((1 - VAL_SPLIT) * x_train.shape[0])
    x_train_sub, x_val = x_train[:num_train], x_train[num_train:]
    y_train_sub, y_val = y_train_cat[:num_train], y_train_cat[num_train:]

    train_gen = datagen.flow(x_train_sub, y_train_sub, batch_size=BATCH_SIZE)

    steps_per_epoch = len(x_train_sub) // BATCH_SIZE

    history = model.fit(
        train_gen,
        steps_per_epoch=steps_per_epoch,
        epochs=EPOCHS,
        validation_data=(x_val, y_val),
        verbose=2
    )

    histories[name] = history

    # Evaluate on test data (no augmentation)
    test_loss, test_acc = model.evaluate(x_test, y_test_cat, verbose=0)
    print(f"Test accuracy ({name}): {test_acc:.4f}")

    # Store basic test results
    results[name] = {
        "test_loss": test_loss,
        "test_acc": test_acc
    }


# ============================================================
# 6. Plot Learning Curves (Loss and Accuracy)
# ============================================================

def plot_metric(histories, metric="loss"):
    plt.figure(figsize=(8, 6))
    for name, hist in histories.items():
        plt.plot(hist.history[metric], label=f"{name} - train")
        plt.plot(hist.history[f"val_{metric}"], linestyle="--", label=f"{name} - val")
    plt.xlabel("Epoch")
    plt.ylabel(metric.capitalize())
    plt.title(f"Training and validation {metric}")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_metric(histories, metric="loss")
plot_metric(histories, metric="accuracy")


# ============================================================
# 7. Bar Chart of Test Accuracy for Each Augmentation
# ============================================================

def plot_test_accuracy_bar(results):
    names = list(results.keys())
    accs = [results[n]["test_acc"] for n in names]

    plt.figure(figsize=(7, 5))
    bars = plt.bar(names, accs)
    plt.ylabel("Test Accuracy")
    plt.title("Test Accuracy by Augmentation Strategy")
    plt.ylim(min(accs) - 0.02, 1.0)
    for bar, acc in zip(bars, accs):
        plt.text(bar.get_x() + bar.get_width() / 2, acc - 0.01,
                 f"{acc:.3f}", ha="center", va="top")
    plt.tight_layout()
    plt.show()

plot_test_accuracy_bar(results)


# ============================================================
# 8. Confusion Matrices and Classification Reports
#    (Focus on Selected Strategies, e.g., baseline vs heavy)
# ============================================================

def plot_confusion_matrix(cm, classes, title="Confusion matrix"):
    plt.figure(figsize=(7, 6))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, ha="right")
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black",
                 fontsize=8)

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    plt.show()


# Choose which models to inspect in detail (you can change this list)
detailed_models = ["baseline", "flip", "rotation", "heavy"]

for name in detailed_models:
    print(f"\nClassification report for augmentation: {name}")
    model = models_dict[name]
    y_pred_probs = model.predict(x_test, verbose=0)
    y_pred = np.argmax(y_pred_probs, axis=1)

    acc = accuracy_score(y_test, y_pred)
    print(f"Test accuracy: {acc:.4f}")

    print(classification_report(y_test, y_pred, target_names=class_names, zero_division=0))

    cm = confusion_matrix(y_test, y_pred)
    plot_confusion_matrix(cm, classes=class_names, title=f"Confusion Matrix - {name}")


# ============================================================
# 9. Summary Printout for Report
# ============================================================

print("\n=== SUMMARY OF TEST ACCURACIES ===")
for name, res in results.items():
    print(f"{name:10s}: {res['test_acc']:.4f}")
```
